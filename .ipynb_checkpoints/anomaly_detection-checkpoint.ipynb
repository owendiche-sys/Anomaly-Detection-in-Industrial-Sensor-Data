## Introduction

This project focuses on detecting anomalies in industrial sensor data. The dataset contains time-stamped sensor readings from various machines, which can indicate normal operation or potential faults.

The goal of this project is to:

- Clean and structure the dataset for analysis  
- Explore sensor readings and identify patterns  
- Detect anomalous behaviour using unsupervised machine learning  
- Visualise anomalies for further investigation  

Anomaly detection is a critical task in industrial monitoring, predictive maintenance, and fault prevention. By identifying unusual sensor readings, companies can prevent downtime, reduce maintenance costs, and improve operational efficiency.

## Objectives

- Split the combined dataset column into `Timestamp`, `SensorId`, and `Value`  
- Pivot the dataset into a format suitable for analysis  
- Perform exploratory data analysis (EDA)  
- Train an Isolation Forest model to detect anomalies  
- Visualise anomalies across sensors and over time  
- Summarise findings and provide insights for predictive maintenance


# Import Required Libraries
We import libraries for:
- data manipulation (`pandas`, `numpy`)  
- visualisation (`matplotlib`, `seaborn`)  
- anomaly detection and evaluation (`scikit-learn`)  


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

# Load Dataset
We load the industrial sensor dataset saved as `data`.  
This dataset should contain sensor readings for industrial machines over time.


```python
df = pd.read_csv("data.csv")  
df.head()
```

# Split Combined Column into Timestamp, SensorId, and Value
Separate the single column into three useful columns for analysis.


```python
# Identify the combined column
combined_col = df.columns[0]

# Split by semicolon
df[['Timestamp', 'SensorId', 'Value']] = df[combined_col].str.split(';', expand=True)

# Drop the original combined column
df = df.drop(columns=[combined_col])

# Convert types
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

df.head()
```

# Pivot Data
Convert from long format to wide format: one row per timestamp, columns for each sensor.


```python
df_wide = df.pivot(index='Timestamp', columns='SensorId', values='Value').reset_index()
df_wide.head()
```

# Create Results Folder for Figures


```python
results_dir = "Results"
if not os.path.exists(results_dir):
    os.makedirs(results_dir)

def save_fig(filename):
    """
    Save the current matplotlib figure to Results folder
    """
    filepath = os.path.join(results_dir, filename)
    plt.savefig(filepath, dpi=300, bbox_inches='tight')
    print(f"Saved figure: {filepath}")
```

#  Exploratory Data Analysis
Visualise distributions and correlations across sensors.


### Histograms

```python
if df_wide.index.name == 'Timestamp':
    df_wide = df_wide.reset_index()

numeric_cols = df_wide.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [col for col in numeric_cols if col != 'Timestamp']

df_wide[numeric_cols].hist(figsize=(12,8), bins=20)
plt.tight_layout()
plt.xlabel(col, fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(fontsize= 12)
plt.yticks(fontsize= 12)
save_fig("sensor_histograms.png")
plt.show()
```

# Data Preprocessing
Scale features for anomaly detection.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_wide[numeric_cols])
```

# Train Isolation Forest
Detect anomalies in the sensor data.


```python
iso_forest = IsolationForest(n_estimators=200, contamination=0.05, random_state=42)
iso_forest.fit(X_scaled)

# Add anomaly column: -1 = anomaly, 1 = normal
df_wide['anomaly'] = iso_forest.predict(X_scaled)
```

### Visualise Anomalies

```python
# Anomaly Count
sns.countplot(x='anomaly', data=df_wide)
plt.title("Anomaly vs Normal Count")
save_fig("anomaly_count.png")
plt.show()
```

```python
num_anomalies = (df_wide['anomaly'] == -1).sum()
total = df_wide.shape[0]
print(f"Number of anomalies detected: {num_anomalies}")
print(f"Proportion of anomalies: {num_anomalies/total:.2%}")
```
